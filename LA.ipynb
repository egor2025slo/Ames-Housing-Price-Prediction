```python
# Ames Housing: SalePrice prediction (regression)
# Goal: build an end-to-end, leakage-safe ML pipeline and evaluate on a held-out test set.


import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.linear_model import Ridge
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# Reproducibility + basic configuration (paths, target, random seed)
sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 140)

RANDOM_STATE = 42
TARGET = "SalePrice"

DATA_PATH = Path("YOUR_PATH")


# Load dataset and run a quick sanity check (shape, columns, types, missing values)
df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
display(df.head())

print("\nInfo:")
df.info()

print("\nMissing (top 15):")
missing = df.isna().sum().sort_values(ascending=False)
display(missing.head(15))


# EDA: price is right-skewed; log-transform stabilizes variance and reduces outlier impact
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df[TARGET], bins=40, kde=True, ax=ax)
ax.set_title("Distribution of SalePrice")
ax.set_xlabel("SalePrice")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(10, 4))
sns.boxplot(x=df[TARGET], ax=ax)
ax.set_title("SalePrice Boxplot")
plt.tight_layout()
plt.show()

df["log_price"] = np.log(df[TARGET])

fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df["log_price"], bins=40, kde=True, ax=ax)
ax.set_title("Distribution of log(SalePrice)")
ax.set_xlabel("log(SalePrice)")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()


# cleaning
no_object_cols = [
    "Alley", "Pool QC", "Fence", "Misc Feature", "Fireplace Qu",
    "Garage Type", "Garage Finish", "Garage Qual", "Garage Cond",
    "Bsmt Qual", "Bsmt Cond", "Bsmt Exposure", "BsmtFin Type 1", "BsmtFin Type 2"
]

df_clean = df.copy()

# Many NaNs are "structural missingness" (feature not present) rather than unknown values -> encode as "None"
for col in no_object_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna("None")

# Drop obvious IDs
id_cols = [c for c in ["Order", "PID"] if c in df_clean.columns]
df_clean.drop(columns=id_cols, inplace=True)

# feature engineering
df_fe = df_clean.copy()

# Age features
df_fe["house_age"] = (df_fe["Yr Sold"] - df_fe["Year Built"]).clip(lower=0)
df_fe["remod_age"] = (df_fe["Yr Sold"] - df_fe["Year Remod/Add"]).clip(lower=0)

# Area features
df_fe["total_house_area"] = df_fe["Gr Liv Area"] + df_fe["Total Bsmt SF"].fillna(0)

# Finished basement area
finished_cols = [c for c in ["BsmtFin SF 1", "BsmtFin SF 2"] if c in df_fe.columns]
df_fe["finished_bsmt_area"] = df_fe[finished_cols].sum(axis=1) if finished_cols else 0
df_fe["total_finished_area"] = df_fe["Gr Liv Area"] + df_fe["finished_bsmt_area"]

# Quality x size interaction
df_fe["quality_area"] = df_fe["Overall Qual"] * df_fe["Gr Liv Area"]

# Bathrooms total (weighted)
for c in ["Full Bath", "Half Bath", "Bsmt Full Bath", "Bsmt Half Bath"]:
    if c not in df_fe.columns:
        df_fe[c] = 0

df_fe["bathrooms_total"] = (
    df_fe["Full Bath"] +
    0.5 * df_fe["Half Bath"] +
    df_fe["Bsmt Full Bath"] +
    0.5 * df_fe["Bsmt Half Bath"]
)

# Binary flags
df_fe["has_pool"] = (df_fe.get("Pool Area", 0) > 0).astype(int)
df_fe["has_fireplace"] = (df_fe.get("Fireplaces", 0) > 0).astype(int)
if "Alley" in df_fe.columns:
    df_fe["has_alley"] = (df_fe["Alley"].astype(str) != "None").astype(int)
else:
    df_fe["has_alley"] = 0


# split + preprocessing
X = df_fe.drop(columns=[TARGET, "log_price"])
y = df_fe["log_price"]

numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = X.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

# train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

# Numeric preprocessing: median imputation + scaling (scaling mainly helps linear models)
numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

# Categorical preprocessing: impute + one-hot encode; ignore unseen categories at test time
categorical_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# ColumnTransformer applies the right preprocessing to each column group in a single, reproducible pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_pipeline, numeric_cols),
        ("cat", categorical_pipeline, categorical_cols)
    ]
)


# Baseline model: Ridge is a strong, interpretable linear baseline with L2 regularization
ridge_pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", Ridge())
])

ridge_grid = GridSearchCV(
    ridge_pipe,
    {"model__alpha": [0.1, 1, 10, 50, 100]},
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)
ridge_grid.fit(X_train, y_train)
best_ridge = ridge_grid.best_estimator_

# Strong model: Gradient Boosting captures non-linearities and interactions beyond linear models
gbr_pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", GradientBoostingRegressor(random_state=RANDOM_STATE))
])

gbr_grid = GridSearchCV(
    gbr_pipe,
    {
        "model__n_estimators": [300, 600],
        "model__learning_rate": [0.03, 0.05, 0.1],
        "model__max_depth": [2, 3]
    },
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)
gbr_grid.fit(X_train, y_train)
best_gbr = gbr_grid.best_estimator_

print("Best Ridge:", ridge_grid.best_params_, "CV RMSE:", -ridge_grid.best_score_)
print("Best GBR:", gbr_grid.best_params_, "CV RMSE:", -gbr_grid.best_score_)


# metric evaluation
def regression_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2

ridge_pred = best_ridge.predict(X_test)
gbr_pred = best_gbr.predict(X_test)

ridge_mae, ridge_rmse, ridge_r2 = regression_metrics(y_test, ridge_pred)
gbr_mae, gbr_rmse, gbr_r2 = regression_metrics(y_test, gbr_pred)

results_log = pd.DataFrame({
    "Model": ["Ridge", "Gradient Boosting"],
    "MAE (log)": [ridge_mae, gbr_mae],
    "RMSE (log)": [ridge_rmse, gbr_rmse],
    "R2": [ridge_r2, gbr_r2]
})
display(results_log)

# Convert predictions back to dollars to report MAE/RMSE in real units
y_test_price = np.exp(y_test)
ridge_pred_price = np.exp(ridge_pred)
gbr_pred_price = np.exp(gbr_pred)

ridge_mae_p, ridge_rmse_p, ridge_r2_p = regression_metrics(y_test_price, ridge_pred_price)
gbr_mae_p, gbr_rmse_p, gbr_r2_p = regression_metrics(y_test_price, gbr_pred_price)

results_price = pd.DataFrame({
    "Model": ["Ridge", "Gradient Boosting"],
    "MAE ($)": [ridge_mae_p, gbr_mae_p],
    "RMSE ($)": [ridge_rmse_p, gbr_rmse_p],
    "R2": [ridge_r2_p, gbr_r2_p]
})
display(results_price)


# diagnostic plots + error analysis (GBR)
y_true = y_test
y_pred = gbr_pred
residuals = y_true - y_pred

fig, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, ax=ax)
mn = min(y_true.min(), y_pred.min())
mx = max(y_true.max(), y_pred.max())
ax.plot([mn, mx], [mn, mx], linestyle="--")
ax.set_title("True vs Predicted (GBR, log-space)")
ax.set_xlabel("True log_price")
ax.set_ylabel("Pred log_price")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(7, 4))
sns.scatterplot(x=y_pred, y=residuals, alpha=0.6, ax=ax)
ax.axhline(0, linestyle="--")
ax.set_title("Residuals vs Predicted (GBR, log-space)")
ax.set_xlabel("Pred log_price")
ax.set_ylabel("Residual (true - pred)")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(7, 4))
sns.histplot(residuals, bins=40, kde=True, ax=ax)
ax.set_title("Residuals distribution (GBR, log-space)")
ax.set_xlabel("Residual")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()

error_df = pd.DataFrame({"true": y_true, "pred": y_pred, "abs_error": np.abs(residuals)})
error_df["price_segment"] = pd.qcut(y_true, q=5, labels=["very_low", "low", "medium", "high", "very_high"])
display(error_df.groupby("price_segment")["abs_error"].mean())


# interpretability: feature importances (GBR)
fitted_preprocessor = best_gbr.named_steps["preprocess"]
ohe = fitted_preprocessor.named_transformers_["cat"].named_steps["onehot"]

cat_features = ohe.get_feature_names_out(categorical_cols)
feature_names = np.concatenate([numeric_cols, cat_features])

importances = best_gbr.named_steps["model"].feature_importances_

importance_df = (
    pd.DataFrame({"feature": feature_names, "importance": importances})
    .sort_values("importance", ascending=False)
)

display(importance_df.head(20))

fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(data=importance_df.head(20), x="importance", y="feature", ax=ax)
ax.set_title("Top Feature Importances (GBR)")
plt.tight_layout()
plt.show()

# End of project.
# See README for full details.
```
